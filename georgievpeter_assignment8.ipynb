<h4>Story Idea #1</h4>

Looking at the institutions with the most robberies, there is one anomaly, which stands out.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv('campus_crime.csv')

data["total_count"] = data["murder"] + data["manslaughter"] + data["forcible"]\
  + data["nonforce"] + data["robbery"] + data["agg_aslt"] + data["burglary"]\
  + data['vehicle'] + data["arson"]

data.sort_values('robbery', ascending=False)[:15]
```

Despite having a total enrolment of just over 20,000, <b>Instituto de Banca y Comercio (IBC) Inc has registered 19 robberies</b>. Arizona State and UCF - second and third in this ranking, have vastly more students but still almost two times fewer robberies than IBC. 

Also, 19 robberies is almost <b>80% of all campus crimes</b> for that institution. 

My first story idea is to investigate IBC and find an explanation for those numbers.

<h4>Story Idea #3</h4>

Immediately after opening the spreadsheet, I wanted to calculate the crime rate per 10,000 students for every institution, i.e. to divide the total_count by the total_enr column and multiply by 10,000. Here's the result:

```python
data["crime_rate"] = (data["total_count"]/data["total_enr"])*10000

data.sort_values('crime_rate', ascending=False)[:10]
```

On the top of this table appears Stony Brook University (SBU) with <b>51 crimes per 10,000 students</b>. This in itself is an interesting story topic but one could also dig into why burglaries are about 80% of all crimes at SBU.
